<span style ="font-size:18;"> Analyze your business's financial performance by cross-referencing your sales data with your internal menu data 
to figure out revenues and costs for the year.

This is done by reading in the menu and sales data into lists,
and outputting the calculations to a dictionary displayed in output.txt </span>

```python
# Import the pathlib and csv library
from pathlib import Path
import csv

# Initialize variables
menu = []
sales = []
report = {}
menu_item = ""
sales_item = ""
quantity = 0  
price = 0.0
cost = 0.0
```

<span style ="font-size:18;"> Read in menu_data.csv and set its contents to a separate list object</span>

```python
#Use a with statement and open the menu_data.csv by using its file path
csvpath = '../Instructions/PyRamen/Resources/menu_data.csv'
with open(csvpath, 'r') as csvfile:
    #Use the reader function from the csv library to begin reading menu_data.csv
    csvreader = csv.reader(csvfile, delimiter=',')

    #Use the next function to skip the header (first row of the CSV).
    header = next(csvreader)

    #Loop over the rest of the rows and append every row to the menu list object (the outcome will be a list of lists)
    for row in csvreader:
        menu.append(row)
```

<span style ="font-size:18;"> Set up the same process to read in sales_data.csv </span>

```python
#Use a with statement and open the menu_data.csv by using its file path
csvpath = '/Users/akanksha/Rutgers/Rutgers-FinTech/02-Homework/02-Python/Instructions/PyRamen/Resources/sales_data.csv'
with open(csvpath, 'r') as csvfile:
    #Use the reader function from the csv library to begin reading sales_data.csv
    csvreader = csv.reader(csvfile, delimiter=',')

    #Use the next function to skip the header (first row of the CSV).
    header = next(csvreader)

    #Loop over the rest of the rows and append every row to the sales list object 
    for row in csvreader:
        sales.append(row)
```

<span style ="font-size:18;"> For each row of the sales data, set the following columns of the sales data to their own variables:
* Quantity
* Menu_Item
</span>

```python
for sale in sales:
    quantity = int(sale[3])
    sales_item = sale[4]
    if sales_item not in report:
        report[sales_item] = {"01-count": 0, "02-revenue": 0, "03-cogs": 0, "04-profit": 0}
        
        '''
        For each row of the menu data, set the following columns of the menu data to their own variables:
            * Item
            * Price
            * Cost
        '''
        for m in menu:
            menu_item = m[0]
            price = float(m[3])
            cost = float(m[4])

            #calculate the profit for each item
            profit = price - cost

            #If the sales_item in sales is equal to the item in menu..
            if sales_item == menu_item:
                #Cumulatively add the values to the corresponding metrics in the report
                report[sales_item]["01-count"] += quantity
                report[sales_item]["02-revenue"] += price * quantity
                report[sales_item]["03-cogs"] += cost * quantity
                report[sales_item]["04-profit"] += profit * quantity
            else: 
                print (f"{sales_item} does not equal {menu_item}! NO MATCH!")
```

<span style ="font-size:18;"> Set up output file </span>

```python
# Set the output file path
output_path = Path('output.txt')

# Open the output path as a file object
with open(output_path, 'w') as txtfile:
    #Output each ramen type as the keys and 01-count, 02-revenue, 03-cogs, and 04-profit metrics as the values for every ramen type
    for key in report:
        txtfile.write(f"{key} : {report[key]}\n")
```
